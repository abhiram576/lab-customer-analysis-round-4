```python
correlations_matrix = df.corr()
correlations_matrix
```

```python
sns.heatmap(correlations_matrix, annot=True)
plt.show()
```

```python
# High correlation pairs  with correlation greater than 0.9
high_correlation_pairs = set()
for i in range(len(correlations_matrix.columns)):
    for j in range(i):
        
        if abs(correlations_matrix.iloc[i, j]) > 0.9:
            col_i = correlations_matrix.columns[i]
            col_j = correlations_matrix.columns[j]
            high_correlation_pairs.add((col_i, col_j))
plt.show()
```
